```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

booksDF = pd.read_csv("C:\\Users\\Khazbin\\Desktop\\Data Analysis\\data-viz-with-python-main\\books.csv", error_bad_lines=False)

years = []

for y in booksDF['publication_date']:
    date = y.split("/")
    years.append(date[2])

booksDF["year"] = years

list(booksDF)
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
booksDF.plot(x="ratings_count",y="text_reviews_count",style="o")
plt.title("Number of Text Reviews vs Number of Ratings", color = 'black')
plt.xlabel("Number of Ratings", color = 'black')
plt.ylabel("Number of Text Reviews", color = 'black')

plt.show()
```

```python
# Use your Python skills to create a visualization here.
barChart = sb.barplot(data=booksDF,x="text_reviews_count",y="ratings_count",color="r",orient="h")
barChart.set(xlabel="Number of Text Reviews",ylabel="Number of Ratings",title="Number of Text Reviews and Ratings")

# I don't have time right now, but given the amount of ratings, I would group that data into categories for both graphs
# to display instead, such as 0-10,000, 10,001-20,000, etc.
```
